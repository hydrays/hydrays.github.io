# 函数极限: $\dfrac{\sin x}{x}$ 的极限

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Step 2: 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
```

## 实验3：$\sin x/x$ 的极限

重要极限：$\displaystyle \lim_{x \to 0} \frac{\sin x}{x} = 1$

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_sinc = np.sin(x) / x
y_sinc[np.abs(x) < 1e-10] = 1  # 处理x=0处的间断

# 可视化
plt.figure(figsize=(12, 6))
plt.plot(x, y_sinc, 'b-', linewidth=2, label='sin(x)/x')
plt.axhline(y=1, color='r', linestyle='--', label='y=1')
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-0.5, 1.1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('sin(x)/x 函数图像与极限行为')
plt.legend()
plt.grid(True)
```

```python

# 放大观察极限点
plt.figure(figsize=(10, 6))
x_zoom = np.linspace(-0.5, 0.5, 500)
y_zoom = np.sin(x_zoom) / x_zoom
y_zoom[np.abs(x_zoom) < 1e-10] = 1
plt.plot(x_zoom, y_zoom, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--')
plt.scatter(0, 1, c='red', s=100, zorder=5, label='极限点 (0,1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('x→0时的局部放大图')
plt.legend()
plt.grid(True)
plt.show()
```

**运行结果与思考**：
1. 函数在 $x=0$ 处看似"无定义"，但极限存在
2. 当 $x\to0$ 时，函数值 $\to 1$，形成"可去间断点"
3. 函数的极限 $\ne$ 该点函数值
