
## 实验4：震荡函数 $\displaystyle\sin(1/x)$ 极限不存在案例

函数 $\displaystyle f(x) = \sin\left(\frac{1}{x}\right)$ 在 $x\to0$ 时无限震荡，极限不存在

```python
x = np.linspace(-1, 1, 5000)  # 高密度采样
x = x[x != 0]  # 移除零点
y = np.sin(1/x)

# 主图展示
plt.figure(figsize=(12, 6))
plt.plot(x, y, 'b-', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('sin(1/x) 震荡函数 (x≠0)')
plt.axvline(x=0, color='r', linestyle='--', label='x=0')
plt.xlim(-0.2, 0.2)  # 放大关键区域
plt.ylim(-1.5, 1.5)
plt.grid(True)

# 插入全局视角小图
ax_inset = plt.axes([0.2, 0.6, 0.2, 0.2])
ax_inset.plot(x, y, 'b-', linewidth=0.3)
ax_inset.set_xlim(-1, 1)
ax_inset.set_ylim(-1.5, 1.5)
ax_inset.set_title('全局视图')
ax_inset.grid(True)

plt.show()
```

```python
# 展示不同逼近路径
x1 = 1/(np.arange(1, 100)*np.pi)   # 路径1: x=1/(nπ)
x2 = 2/(np.arange(1, 100)*np.pi)   # 路径2: x=2/(nπ)
y1 = np.sin(1/x1)
y2 = np.sin(1/x2)

plt.figure(figsize=(10, 6))
plt.plot(x1, y1, 'ro-', markersize=4, label='x=1/(nπ) → f(x)=0')
plt.plot(x2, y2, 'bs-', markersize=4, label='x=2/(nπ) → f(x)=1')
plt.axvline(x=0, color='k', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('不同逼近路径得到不同极限值')
plt.legend()
plt.grid(True)
plt.show()
```

**运行结果与思考**：
1. 在x=0附近无限震荡，振幅不变
2. 沿不同路径逼近x=0：
   - 当 $x_n = \frac{1}{n\pi}$ 时，$f(x_n) = 0$
   - 当 $x_n = \frac{2}{(4n+1)\pi}$ 时，$f(x_n) = 1$
3. 极限不存在！因为不同路径得到不同结果
