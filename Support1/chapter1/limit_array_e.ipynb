# 数列极限: 复利的计算与欧拉数e

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Step 2: 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
```


## 实验2：通过计算复利得到自然常数 $e$

$\displaystyle \lim_{n \to \infty} (1 + \frac{1}{n})^n = e$

```python
def compound_interest(n):
    """计算复利"""
    return (1 + 1/n)**n

# 计算不同复利次数下的收益
n_values = np.arange(1, 101)  # 1到100次复利
daily = [compound_interest(k*365) for k in range(1, 101)]  # 按日复利
continuous = [np.exp(1)] * 100  # 连续复利

# 可视化
plt.figure(figsize=(12, 6))
plt.plot(n_values, [compound_interest(n) for n in n_values], 'b-')
plt.axhline(y=np.e, color='r', linestyle='--', label='y = e')
plt.xlabel('复利次数 n')
plt.ylabel('本金增长值')
plt.title('(1+1/n)^n 随n增大逼近e')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 特殊点展示
print(f"年度复利: {compound_interest(1):.2f}")
print(f"月度复利: {compound_interest(12):.4f}")
print(f"每日复利: {compound_interest(365):.6f}")
print(f"连续复利: {np.e:.8f}")
```
